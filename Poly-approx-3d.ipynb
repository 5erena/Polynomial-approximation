```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import axes3d  
from mpl_toolkits.mplot3d.axes3d import *
```

```python
#data points
x = np.array([0,0,1,1,2,2])
y = np.array([0,2,0,2,1,3])
A = np.ones((6,6))
b = np.ones((6,1))
for i in range(6):
    A[i][1] = x[i]
    A[i][2] = y[i]
    A[i][3] = x[i]*y[i]
    A[i][4] = x[i]**2
    A[i][5] = y[i]**2
    b[i][0] = np.exp(x)[i]*np.sin(y[i])
c = np.linalg.solve(A,b)
print(c)
```

```python
def f(x,y):
    return np.exp(x)*np.sin(y)
def p(x,y):
    return c[0]+c[1]*x+c[2]*y+c[3]*x*y+c[4]*x**2+c[5]*y**2

x = np.linspace(-1, 3, 40)
y = np.linspace(-1, 3, 40)

X, Y = np.meshgrid(x,y)
F = f(X,Y)
P = p(X,Y)

fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(X, Y, F, rstride=1, cstride=1, cmap='viridis', edgecolor='none') #plot f
ax.plot_surface(X, Y, P, rstride=1, cstride=1, edgecolor='none') #plot p
ax.set_title('Exact function vs polynomial approximation')
ax.set_xlabel('$X$')
ax.set_ylabel('$Y$')
ax.set_zlabel('$Z$')
#ax.view_init(5, 265)
plt.savefig('hw7-2.pdf')
```
